```python
# Which week was the most busiest 
df.groupby('day').count()['age'].plot(kind='bar')
```

```python
def which_week(x):
  if x <= 7:
    return 1
  elif x>7 and x<=14:
    return 2
  elif x>14 and x<=21:
    return 3
  elif x>21 and x<=28:
    return 4
  else:
    return 5    
```

```python
which_week(24)
```

```python
df['week'] = df['day'].apply(which_week)
```

```python
df.groupby('week').count()['age'].plot(kind='bar')
```
